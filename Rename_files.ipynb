this code is for rename the files if accidently name the iamge with the same name.

```python
import os
import shutil

# Define your source and destination directories.
source_dir = r"A:\_Ongoing\20250221_ActRep"
dest_dir = r"A:\_Ongoing\20250221_ActRep_rename"

# Create the destination directory if it does not exist.
if not os.path.exists(dest_dir):
    os.makedirs(dest_dir)
    print(f"Created directory: {dest_dir}")

# Loop through all files in the source directory.
for filename in os.listdir(source_dir):
    # Process only .nd2 files.
    if filename.endswith(".nd2"):
        src_file_path = os.path.join(source_dir, filename)
        
        # Split the filename into base name and extension.
        base, ext = os.path.splitext(filename)
        parts = base.split('_')
        
        # Check if the file is a "Rep" file with an extra numeric suffix.
        # For example, "Rep_A1_001_0001.nd2" splits into ['Rep', 'A1', '001', '0001']
        if parts[0] == "Rep" and len(parts) >= 4:
            # Construct the new filename as "Act_<part2>_<part3>" and retain the extension.
            new_filename = f"Act_{parts[1]}_{parts[2]}{ext}"
        else:
            # If the file doesn't match the renaming pattern, leave its name unchanged.
            new_filename = filename
        
        dest_file_path = os.path.join(dest_dir, new_filename)
        
        # To avoid any name conflicts in the destination, check if the file already exists.
        if os.path.exists(dest_file_path):
            base_new, ext_new = os.path.splitext(new_filename)
            i = 1
            while os.path.exists(os.path.join(dest_dir, f"{base_new}_{i}{ext_new}")):
                i += 1
            new_filename = f"{base_new}_{i}{ext_new}"
            dest_file_path = os.path.join(dest_dir, new_filename)
        
        # Move the file from the source to the destination folder.
        shutil.move(src_file_path, dest_file_path)
        print(f"Moved: {filename} -> {new_filename}")
```
